Question: Does hours of work affect the hours of sleep for a employee who works remotely?

Import Pandas, Seaborn, and matplotlin.pyplot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Read CSV dataset and set it to df

```python
df=pd.read_csv("Dataset Folder/wfh_burnout_dataset.csv")
df = df.sort_values("sleep_hours", ascending=True)
df.head(20)
```

Checking Data to make sure there is no null or duplicates.

```python
df.isna().sum()
```

```python
df["user_id"].duplicated().sum()
```

Since there is 150 unique workers 2000-150=1850

Relation plot

```python
sns.regplot(data=df, x="work_hours", y="sleep_hours", scatter=True, scatter_kws={"zorder": 1},line_kws={"zorder": 2, "color":"black"}, label="Sleep hours vs Work Hours")
plt.xlabel("Work Hours")
plt.ylabel("Sleep Hours")
plt.title("Sleep Hours vs Work Hours")
```

**Add a new variable that calculates the non active screen time spent during work**

```python
df["non_active_hours"] = (df["screen_time_hours"] - df["work_hours"])*-1
```

```python
#Round work hours to the nearest half number.
df["work_hours"] = round(df["work_hours"]*2)/2
```

```python
sns.barplot(data=df,x="work_hours",y="non_active_hours",color="#CB1E1E",errcolor="#000000")
plt.ylabel("Non Active Hours")
plt.xlabel("Work Hours")
plt.title("Work Hours vs Non Active Hours")
plt.xticks(rotation=45)
```

**The more hours a worker works the less time they spend not actively looking at a screen**

```python
df["sleep_hours"] = round(df["sleep_hours"]*2)/2
```

```python
sns.countplot(data=df,x="sleep_hours")
plt.title("Total Count of Hours Slept")
plt.xlabel("Sleep Hours")
plt.ylabel("Total Amount")
```
